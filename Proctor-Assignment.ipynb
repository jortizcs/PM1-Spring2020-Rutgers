```python
import csv
import sys
import random
import numpy as np
import os
```

```python
def in_lab(sections_str):
    sections = sections_str.split(',')
    for s in sections:
        if ':254:' in s:
            return True
    return False
```

```python
def in_lec(sections_str):
    sections = sections_str.split(',')
    for s in sections:
        if ':252:' in s:
            return True
    return False
```

```python
def assign_proc(proctors, roster_file='PM1-spring2020-Jan26.csv', res_file=None, ):

    file_=None
    if res_file is not None:
        file_ = open("lec_sections.csv","w")
    proctor_keys=np.asarray(list(proctors.keys()))
    protctor_assign_count = np.zeros(len(proctors.keys()))
    #print (proctor_keys)
    pidx =0
    
    for pk in proctors.keys():
        proctor_fn = "-".join(pk.split())+'-'+str(len(proctors.keys()))+'.csv'
        if os.path.exists(proctor_fn):
            os.remove(proctor_fn)

    with open(roster_file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        c=0
        for row in csv_reader:
            if line_count == 0:
                #print(f'Column names are {", ".join(row)}')
                #print (", ".join(row))
                line_count += 1
                #print(",".join([row[0], row[1], row[2],row[3]]))
                if file_ is not None:
                    file_.write(",".join([row[0], row[1], row[2],row[3]])+"\n")
            else:
                #print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
                #2020:1:14:332:252:01:,2020:1:14:332:254:01:
                if row[4] == 'Student' and in_lec(row[2]) and row[5]=='active':
                    section_id=''
                    for v in row[2].split(","):
                        if ':252:' in v:
                            section_id = v.split(':')[5]
                    name=row[0]
                    netid=row[1]
                    #pidx = random.randint(0,len(proctors.keys())-1)
                    #pidx = round(random.uniform(0,len(proctors.keys())-1))
                    pk = proctor_keys[pidx]
                    protctor_assign_count[pidx]+=1
                    pidx=(pidx+1)%len(proctors.keys())
                    proctor_link=proctors[pk]
                    proctor_fn = "-".join(pk.split())+'-'+str(len(proctors.keys()))+'.csv'                    
                    proctor_file = open(proctor_fn,'a+')
                    proctor_file.write(netid+'@scarletmail.rutgers.edu'+'\n')
                    proctor_file.close()
                    
                    #print(",".join([row[0], row[1], section_id,row[3]]))
                    print(name, '\t'+netid+'@scarletmail.rutgers.edu', '\t'+proctor_link)
                    if file_ is not None:
                        file_.write(",".join(["\""+row[0]+"\"",row[1], section_id,row[3]])+"\n")
                    c+=1
                line_count += 1

        print(f'Processed {c} lines.')

    if file_ is not None:
        file_.close() 
    print (protctor_assign_count)
```

```python
proctors_={}
proctors_['tahiya chowdhury']='webex_url_tahiya'
proctors_['murtadha aldeer']='webex_url_murtadha'
proctors_['mengmei ye']='webex_url_mengmei'
proctors_['jorge ortiz']= 'webex_url_jorge'
proctors_['chaoji zuo'] ='webex_url_chaoji'
proctors_['archana arjula']='webex_url_archana'

roster='Rutgers-PM1/2020/PM1-roster-3-18-2020.csv'
```

```python
assign_proc(roster_file=roster, proctors=proctors_)
```
